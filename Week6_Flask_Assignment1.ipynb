## Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is lightweight python framework that is used for making small to medium web applications. It is microframework which means that it does not include many features that are found in larger frameworks such as Django.

+ Lightweight and easy to learn :  Flask is very lightweight framework which makes it easy to get started with.

+ Flexible :  Flask is very flexible which means we can use it to build wide range of web applications.

+ Extendable : Flask is very extendable, which means that you can add new features and functionality to application.

+ Community : Flask has large and active community which means that there are many resources available to help you learn and use flask.

## Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask is process of mapping URLs to functions in your application. This allows to control what happens when user visit a particular URL `/` to a function that displays the home page of your application.

We use app routes in Flask to : 
* Control what happens when a user visits a particular URL

* Organize your application code into logical units

* Make application more reusable and scalable

* Improve the performance of your application by caching frequently accessed routes.

* write unit tests for specific routes and functionalities to application

## Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
- Company Name: ABC Corporation 
- Location: India
- Contact Detail: 999-999-9999

```python
from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return "<h1>Welcome to ABC Corporation</h1>"

@app.route('/')
def info():
    return "<h1> Company Name : ABC Corporation <br> Location: India <br> Contact Detail : 999-999-9999</h1>"

if __name__ == '__main__':
    app.run(host = '0.0.0.0')
```

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the `url_for()` function is used to dynamically generate urls for specific routes within the application. It takes name of route function as its first argument and any additional keyword arguments representing route parameters.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return "Welcome to the homepage!"

@app.route('/about')
def about():
    return "This is the about page."

if __name__ == '__main__':
    with app.test_request_context():
        home_url = url_for('home')
        about_url = url_for('about')
        print("Home URL:", home_url)
        print("About URL:", about_url)
```
