## Q1. Explain GET and POST methods.

### GET Method

The GET method is used to request data from a resource in the form of url. When we use browser to visit website or click on link, basically we are making a GET request. GET requests are designed to retrieve data and they are included in the URL as query parameter

for example, if we are searching for laptop, the GET request may look : 

`https://example.com.search?query=laptop`

### POST Method

The POST method is used to send data to server to create or update resources. It is used for submitting forms like user registration or comment submission. The data sent using POST is included in the request body rather than the URL.

POST requests are not idempotent,meaning the same request multiple times can result in different outcomes.

## Q2. Why is request used in Flask?

The Flask Framework provides a built-in 'request' object that allows to access various aspects of incoming request, such as form data, query parameters, headers, etc. It is important to handle and process client requests within flask application.

Request is used in flask for following:
+ Access Data :  `request` object allows to access data sent by the client, such as form data submitted by users, query parameters, JSON payloads, etc

+ Dynamic Routing : With `request` objectm we can extract variables ffrom the url route pattern and use them to customise the behaviour of application.

+ Cookies : The `request` objects allows you to access cookies sent by the client's browser enabling you to maintain the state and user sessions.

## Q3. Why is redirect() used in Flask?

In Flask, the `redirect()` function is used to perform URL redirection within a web application. URL redirection is the process of sending a client's browser to different URL, typically after a certain action has been performed or to ensure that the user accesses the correct page.

Few scenarios where `redirect()` used includes:
+ After Form Submission : When a user submits a form, you might want to redirect them to different page after the form has been successfully processed

+ Changing URLs : If application has multiple URLs that point to similar content, you might want to ensure that all users access the same canonical URL.

```python
from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route('/')
def index():
    # Redirect to the 'about' page
    return redirect(url_for('about'))

@app.route('/about')
def about():
    return "This is the About page."

if __name__ == '__main__':
    app.run()
```

## Q4. What are templates in Flask? Why is the render_template() function used?

Templates in Flask are used to separate the presentation of web application from its logic. In other words it is used to separate HTML from Python code. Templates enable you to create dynamic and interactive web pages by combining HTML structure with placeholders for dynamic content that will be generated by your Flask application.

The `render_template()` function in Flask is used to render these templates and generate HTML pages that are sent to the client's browser. This function takes the name of template file as an argument and any additional variables you want to pass to the template. It then combines the template's structure with data provided to produce complete HTML page.

```python
<!-- templates/index.html -->
<!DOCTYPE html>
<html>
<head>
    <title>{{ title }}</title>
</head>
<body>
    <h1>Welcome to {{ title }}</h1>
    <p>This is a dynamic web page generated with Flask and Jinja2.</p>
</body>
</html>
```

```python
from flask import Flask, render_template

app = Flask(__name__)

@app.route('/')
def index():
    title = "My Flask App"
    return render_template('index.html', title=title)

if __name__ == '__main__':
    app.run()
```
